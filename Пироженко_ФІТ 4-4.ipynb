```python
import pandas as pd
import numpy as np

# Зчитуємо дані з вказаного URL
url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

# Зчитуємо таблицю в DataFrame
data = pd.read_html(url, header=0)[2]  # Обираємо другий елемент списку, де знаходиться таблиця

df_1 = data.head(5)
# Виводимо перші 5 рядків таблиці
print(df_1)

```

```python
# Визначаємо загальну кількість років та місяців у даних
num_years = len(data['Рік']) - 1  # віднімання 1, оскільки перший стовпець - назва
num_months = num_years*12
print("Загальна кількість років =", num_years)
print("Загальна кількість місяців =", num_months)
```

```python
#Визначити кількість рядків і стовпчиків

num_rows, num_columns = data.shape
print(f"Кількість рядків: {num_rows}, Кількість стовпчиків: {num_columns}")
```

```python
#Замінити в таблиці "—" на значення  NaN
#Також заміню назви стовпців на зручні
data = data.rename(columns={
    "Рік": "Year",
    "В національній валюті": "Salary in UAH",
    "В доларах США": "Salary in USD",
    "Кількість кг хліба[20] на середню зарплату": "Bread per Salary",
    "Кількість БігМаків на середню зарплату": "BigMacs per Salary"
})
data = data.replace("—", np.nan)
data
```

```python
# 5. Визначити тип стовпців за допомогою dtypes.
column_types = data.dtypes
print("Типи стовпців:")
print(column_types)
```

```python
# 6. Видаліть передостанній і останній стовпчики
data = data.drop(columns=["Bread per Salary", "BigMacs per Salary"])
data
```

```python
#Видалимо трикутники
data["Salary in UAH"] = data["Salary in UAH"].str.replace('▲', "").str.replace('▼', "")
data["Salary in USD"] = data["Salary in USD"].str.replace('▲', "").str.replace('▼', "")
data
```

```python
data.dtypes
```

```python
# Перетворіть тип даних на числовий
data['Salary in UAH'] = data['Salary in UAH'].str.replace(',', '.').astype(float)
data['Salary in USD'] = data['Salary in USD'].str.replace(',', '.').astype(float)





data
```

```python
#Ще раз перевіримо тип даних

data.dtypes
```

```python
# Розрахунок частки пропусків в кожному стовпці
missing_percentage = (data.isnull().sum() / len(data)) * 100

# Виведення результатів
print("Частка пропусків в кожному стовпці:")
print(missing_percentage)


```

```python
# Заміна пропущених значень на середні по кожному стовпцю
data = data.fillna(data.mean())
```

```python
max_salary_year = data[data['Salary in UAH'] == data['Salary in UAH'].max()]
print("Рік з найвищою зарплатою в Україні в гривнях:", max_salary_year['Year'].values[0])
print("Зарплата:", max_salary_year['Salary in UAH'].values[0])
```

```python
min_salary_usd_year = data[data['Salary in USD'] == data['Salary in USD'].min()]
print("Рік з мінімальною зарплатою в Україні в доларах США:", min_salary_usd_year['Year'].values[0])
print("Зарплата в доларах США:", min_salary_usd_year['Salary in USD'].values[0])
```

```python
above_average_years = data[data['Salary in UAH'] > data['Salary in UAH'].mean()]
print("Роки з зарплатою в гривнях вище середньої:")
print(above_average_years['Year'].tolist())
```

```python
import matplotlib.pyplot as plt

# Діапазон років (видаляємо останній рік 2023)
years = data[data['Year'] < 2023]['Year']

# Зарплата в гривнях (UAH) (видаляємо останній рік 2023)
salary_uah = data[data['Year'] < 2023]['Salary in UAH']

# Округлення зарплати до цілих чисел
rounded_salary_uah = [int(salary) for salary in salary_uah]

# Створення більш детального графіку з більшими відступами між точками
plt.figure(figsize=(12, 8))
plt.plot(years, salary_uah, marker='o', linestyle='-', color='b', linewidth=2, markersize=8, label='Зарплата в гривнях')
plt.title('Зарплата в Україні за роками (без 2023 року)', fontsize=16)
plt.xlabel('Рік', fontsize=14)
plt.ylabel('Зарплата в гривнях (UAH)', fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)

# Збільшення відступів між точками на осі X
plt.xticks(years, [str(year) for year in years], rotation=45, fontsize=10, ha='right')

# Додаємо відповідні значення округленої зарплати (цілі числа) біля лінії на графіку
for i, (year, salary) in enumerate(zip(years, rounded_salary_uah)):
    plt.text(year, salary, f'{salary}', ha='center', va='bottom', fontsize=6.5)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Виберіть дані за всі роки окрім останнього (2023 рік)
filtered_data = data[data['Year'] < 2023]

# Групуємо дані за роками і обчислюємо середню зарплату
average_salary_by_year = filtered_data.groupby('Year')['Salary in UAH'].mean()

# Створюємо графік середньої зарплати в розрізі років
plt.figure(figsize=(12, 6))
plt.plot(average_salary_by_year.index, average_salary_by_year.values, marker='o', linestyle='-', color='b')
plt.title('Середня зарплата в Україні за роками (без 2023 року)', fontsize=16)
plt.xlabel('Рік', fontsize=14)
plt.ylabel('Середня зарплата в гривнях (UAH)', fontsize=14)
plt.grid(True)
plt.xticks(average_salary_by_year.index, rotation=45)
plt.tick_params(axis='both', which='major', labelsize=12)
# Збільшення відступів між точками на осі X
plt.xticks(years, [str(year) for year in years], rotation=45, fontsize=10, ha='right')

# Додаємо відповідні значення округленої зарплати (цілі числа) біля лінії на графіку
for i, (year, salary) in enumerate(zip(years, rounded_salary_uah)):
    plt.text(year, salary, f'{salary}', ha='center', va='bottom', fontsize=6.5)

plt.tight_layout()
plt.show()
```

```python
# Виберіть дані за останні два роки (2021 та 2022 роки)
last_two_years_data = data[(data['Year'] >= 2021) & (data['Year'] <= 2022)]

# Виберіть дані за останні п'ять років (включаючи 2018, 2019, 2020, 2021 та 2022 роки)
last_five_years_data = data[(data['Year'] >= 2018) & (data['Year'] <= 2022)]

# Обчисліть середню зарплату за останні два роки
average_salary_last_two_years = last_two_years_data['Salary in UAH'].mean()

# Обчисліть середню зарплату за останні п'ять років
average_salary_last_five_years = last_five_years_data['Salary in UAH'].mean()

# Порівняйте середні зарплати і визначте різницю
difference = average_salary_last_two_years - average_salary_last_five_years

# Визначте напрямок різниці
if difference > 0:
    direction = "збільшилася"
elif difference < 0:
    direction = "зменшилася"
else:
    direction = "не змінилася"

# Виведіть результати
print(f"Середня зарплата за останні два роки: {average_salary_last_two_years:.2f} гривень")
print(f"Середня зарплата за останні п'ять років: {average_salary_last_five_years:.2f} гривень")
print(f"Різниця: {abs(difference):.2f} гривень ({direction})")
```
